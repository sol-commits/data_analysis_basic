```python
import matplotlib.pyplot as plt
import numpy as np
```

# bar chart

```python
# Data for Bar chart
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [30, 50, 25, 40]
```

```python
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
color = ['blue' if value >30 else 'orange' for value in values]

plt.bar(categories, values, color=color)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

# stacked bar

```python
# Data for 'Stacked' Bar chart
categories_stacked = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values1_stacked = [15, 30, 25, 10]
values2_stacked = [10, 20, 30, 25]
```

```python
plt.bar(categories_stacked, values1_stacked, label='Value 1', color='blue')
plt.bar(categories_stacked, values2_stacked, bottom=values1_stacked, label='Value 2', color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.show()
```

# grouped Bar Chart

```python
# Data for 'Grouped' Bar chart
x_grouped = np.arange(len(categories_stacked))
width = 0.3
```

```python
plt.bar(x_grouped - width/2, values1_stacked, width=width, label='Value 1', color='blue')
plt.bar(x_grouped + width/2, values2_stacked, width=width, label='Value 2', color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(x_grouped, categories_stacked)
plt.legend()
plt.show()
```

# line chart

```python
# Data for Line chart
x_values = [1, 2, 3, 4, 5]
y_values = [10, 20, 15, 25, 30]
```

```python
# Create a line chart
plt.plot(x_values, y_values, marker='o', linestyle='--')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Chart')
plt.grid(True)  # Add grid lines
plt.show()
```

## line chart with a threshold line

```python
plt.plot(x_values, y_values, marker='o', linestyle='-', color='blue', label='customized label')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Customized Line Chart')
plt.legend(loc='upper left')
plt.grid(True)
plt.xlim(0, 6)
plt.ylim(0, 35)
plt.xticks(np.arange(1, 6, step=1))
plt.yticks(np.arange(0, 36, step=5))

# add a horizontal line
plt.axhline(y=20, color='red', linestyle='--', linewidth=2, label='Threshold')
# add text annotation
plt.text(3.5, 22, 'Threshold', color='red', fontsize=12)

plt.show()
```

# multiple charts - bar, line

```python
plt.figure(figsize=(10, 5))

# Subplot 1 - Bar Chart
plt.subplot(1, 2, 1)
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')

# Subplot 2 - Line Chart
plt.subplot(1, 2, 2)
plt.plot(x_values, y_values, marker='o', linestyle='--')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Chart')
plt.grid(True)

plt.tight_layout() # adjust the layout to avoid overlapping
plt.show()
```

# scatter plot

```python
# Data for scatter plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
colors_scatter = np.random.rand(50)
sizes_scatter = np.random.randint(10, 100, 50)
```

```python
# Create a scatter plot
plt.scatter(x_scatter, y_scatter, c=colors_scatter, s=sizes_scatter, alpha=0.7)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')
plt.colorbar(label='Color Scale')
plt.show()
```

# pie chart

```python
# Data for pie chart
labels_pie = ['Category A', 'Category B', 'Category C', 'Category D']
sizes_pie = [30, 15, 25, 30]
```

```python
plt.pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=90) # 시작 각도가 90도
plt.axis('equal') # 가로 세로 비율
plt.title('Pie Chart')
plt.show()
```

# histogram

연속형 데이터에서 데이터의 분포를 확인하는데 용이

```python
# Data for histogram
data_histogram = np.random.randn(1000)
```

```python
# Create a histogram
plt.hist(data_histogram, bins=20, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

# box plot

```python
# Data for box plot
data_boxplot = [np.random.normal(0, std, 100) for std in range(1, 5)]
```

```python
plt.boxplot(data_boxplot, labels=['Box 1', 'Box 2', 'Box 3', 'Box 4'])
plt.xlabel('Box Number')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```

# violin plot

```python
# Data for violin plot
data_violin = [np.random.normal(0, std, 100) for std in range(1, 5)]
```

```python
plt.violinplot(data_violin, showmedians=True)
plt.xlabel('Data Series')
plt.ylabel('Values')
plt.title('Violin Plot')
plt.xticks(np.arange(1, 5), ['Series 1', 'Series 2', 'Series 3', 'Series 4'])
plt.show()
```

# heatmap

복잡한 데이터셋에서 상관관계를 시각화하는데 사용

```python
data_heatmap = np.random.rand(5, 5)
```

```python
data_heatmap
```

```python
plt.imshow(data_heatmap, cmap='coolwarm', interpolation='nearest') # interpolation(보간) 알고 있는 데이터 값을 이용하여 모르는 값을 추정(데이터 간 구분이 모호할 때)
plt.colorbar(label='Color Scale')
plt.title('Heatmap')
plt.show()
```

## add annotations(text) to the heatmap

```python
plt.imshow(data_heatmap, cmap='coolwarm', interpolation='nearest')

# add annoatations
for i in range(data_heatmap.shape[0]):
    for j in range(data_heatmap.shape[1]):
        plt.text(j, i, f'{data_heatmap[i, j]:.2f}', ha='center', va='center')
        
plt.colorbar(label='Color Scale')
plt.title('Heatmap with Annotations')
plt.show()
```
