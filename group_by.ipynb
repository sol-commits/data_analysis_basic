```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'B': ['one', 'two', 'one', 'three', 'three', 'two', 'one', 'three', 'two', 'two'],
    'C': np.random.randn(10),
    'D': np.random.randn(10)
})
```

```python
df.head()
```

```python
df.groupby('A').agg({'C':np.sum, 'D': np.mean})
```

```python
df.groupby('A').agg(
    max_C = pd.NamedAgg(column='C', aggfunc='max'),
    min_D = pd.NamedAgg(column='D', aggfunc='min'),
    mean_C = pd.NamedAgg(column='C', aggfunc='mean'),
)
```

```python
df.groupby('A').filter(lambda x: len(x) > 2)
```

```python
df.groupby('A').filter(lambda x: len(x) > 5)
```

```python
df.groupby('B').filter(lambda x: len(x) > 3)
```

```python
df.groupby('A')['D'].transform('mean')
```

```python
df.head()
```

```python
def normalize(x):
    return (x-x.mean())/x.std()
```

```python
df.groupby('A')['C'].apply(normalize)
```

```python
df.groupby('A')['C'].rank()
```

```python
df.select_dtypes(include='number').groupby(df['A']).agg(lambda x: x.max()-x.min())
```
